## Dataframe from list and dictionary

```python
# list
data = [('1/1/2020',6), ('2/1/2020',7), ('3/1/2020',8), ('4/1/2020',9)]
df2 = pd.DataFrame(data, columns=["Day","Windspeed"])
df2
```

```python
# dictionary
data= {
    "Day":['5/1/2020', '6/1/2020', '7/1/2020', '8/1/2020'],
    'Windspeed': [6,7,8,9]
}
df3 = pd.DataFrame(data)
df3
```

### Concatination of dataframe

```python
df4 = pd.concat([df2, df3], ignore_index=True)
df4
```

# Stock Data Plot

```python
import mplfinance as mpl
```

```python
google = pd.read_csv('GOOG.csv', parse_dates=['Date'])
google['Date'] = pd.to_datetime(google['Date'], errors='coerce')
google.set_index('Date', inplace=True)
google.head()
```

```python
mpl.plot(google, type='candle')
# type -> 'candle', 'line', 'ohlc' -> (moving avg) mav=4
```

```python
mpl.plot(google, type='ohlc', mav=(3,5,7), volume=True, show_nontrading=True)
```

## Standerdization

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler().fit(google)
data = scaler.transform(google)
data
```

```python
# convert data to dataframe
standardScaler = pd.DataFrame(data, index = google.index, columns = google.columns)
standardScaler
```
